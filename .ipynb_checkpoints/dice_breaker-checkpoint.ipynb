```python
# The purpose of this program is to use the random module in python to roll a dice for our icebreaker today.
# Korren is going to attempt to do this live, be prepared for typos.

# import our random library
import random

# declare our varibles to determine the smallest and largest values we want selected
min = 1
max = 6

# declare a variable to keep track of if we need to roll again
roll_again = "yes"

# start our loop
while roll_again == "yes":
    # print a message to indicate our die is rolling
    print("Rolling the die...")
    
    # declare varible for the die value and assign it a random number
    die_value = (random.randint(min,max))
    
    # print the value of the die
    print("The value of the die is: ", die_value)
    
    # create the if statement t otell the program what to do for each potential number
    if die_value == 1:
        print('If you were invisible for a day what would you choose to do?')
    elif die_value == 2:
        print("If you could choose a unique super power what would it be?")
    elif die_value == 3:
        print("What is the grossest food you have ever tried?")
    elif die_value == 4:
        print("Who is your favourite book or movie character?")
    elif die_value == 5:
        print("What is your favourite game to play?")
    else:
        print("What is one thing you are really good at?")
    
    # let the program know whether to roll again
    roll_again = input("Roll again? yes/no")
    
# Message to show program is ended
print("Program has ended")
```
